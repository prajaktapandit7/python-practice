```python
#A4: Write a program to aid Josephus in figuring out where to stand in the circle


class Soldier:                              #soldier class parameters
    def __init__(self,data):                #initializing soldier object
        self.data = data
        self.next = None
        self.prev = None

    def insert_after(x, data):
        y = Soldier(data)                   #make a new soldier object
        z = x.next                          #y goes between x and z
        y.prev = x
        y.next = z
        x.next = y
        z.prev = y

    def kill(self):                         #function to kill soldier and adjust links of previous and next soldier
        x = self.data
        self.prev.next = self.next          #link previous and next nodes to each other
        self.next.prev = self.prev
        print ("\nSoldier ", x, " bites the dust")

```

```python
class Army:
    def __init__(self, n):                  #initializing Army class           
        
        self.victim =  Soldier(str(1))      #creating Soldier 1
        self.victim.next = self.victim
        self.victim.prev = self.victim
        x = n 
        
        while (1 < x <= n):                 #adding more soldiers based on user-specified count
            self.victim.insert_after(str(x))
            x = x - 1     
    
    def advance(self, k):                   #function to traverse through the DLL based on the user-specified spacing
        n = self.victim.prev
        
        print("\nAdvancing (to the tune of Queen's Another one bites the dust): ")
        while(n.data != None):              #loop to continue moving through the circle of soldiers
            i = 1
            while i <= k:                   #loop to advance by the given spacing and kill soldiers 
                n = n.next
                i = i + 1               
            n.kill()                        #calling the kill function
            
            if n.next == n.prev:            #exit condition when only one soldier is left
                print ("\nThe last remaining soldier is ", n.next.data)
                break

    def kill_all(self):
        k = int(input("\nEnter spacing between victims, between 1 and n: "))
        print("\nSpacing: ", k)
        self.advance(k)                     #calling the advance function and passing the spacing provided by user

```

```python
n = int(input("\nEnter n number of soldiers, at least 2: "))
print("\nNumber of soldiers: ", n)

my_army = Army(n)                    #creating Army by passing the number of soldiers entered by user
my_army.kill_all()                   #calling the function to kill soldiers
```
