```python
#NAME: PRAJAKTA PANDIT
#INFO 206A
#ASSIGNMENT 3: Write a Python program that draws "string art"

import drawSvg as draw
d = draw.Drawing(600, 600, origin='center') #set canvas

#Sample co-ordinates

#starting point of stick A
#x1a = 0
#y1a = 0

#ending point of stick A
#x2a = -50
#y2a = 185

#starting point of stick B
#x1b = 100
#y1b = -40

#ending point of stick B
#x2b = 150
#y2b = 145

print ("Co-ordinates for Stick A:")

#starting point of stick A
print ("Enter x co-ordinate of starting point of stick A:") 
x1a = int(input())
print ("Enter y co-ordinate of starting point of stick A:") 
y1a = int(input())

#ending point of stick A
print ("Enter x co-ordinate of ending point of stick A:") 
x2a = int(input())
print ("Enter y co-ordinate of ending point of stick A:") 
y2a = int(input())

print ("\nCo-ordinates for Stick B:")

#starting point of stick B
print ("Enter x co-ordinate of starting point of stick B:") 
x1b = int(input())
print ("Enter y co-ordinate of starting point of stick B:") 
y1b = int(input())

#ending point of stick B
print ("Enter x co-ordinate of ending point of stick B:") 
x2b = int(input())
print ("Enter y co-ordinate of ending point of stick B:") 
y2b = int(input())

d.append(draw.Line(x1a, y1a, x2a, y2a, stroke='red', stroke_width=2, fill='none'))  # stick A
d.append(draw.Line(x1b, y1b, x2b, y2b, stroke='red', stroke_width=2, fill='none'))  # stick B

n = 1 #number of strings

#accept input to define # of strings
while n<=1:
    print ("Enter number of strings:") 
    n = int(input())
    if n<=1:
        print ("Number of strings must be greater than 1!")
    print ("You entered:", n)

#error handling for divide by zero error
try:
    check = 1/(n-1)
    
except ZeroDivisionError as e:
    print("Error: Cannot divide by zero")

f = 0 #fraction of length of stick

#draw strings
while f <= 1:
    color = '#%02x%02x%02x' % (22 , 229- int(f*150), 77+ int(f*30)) #varying color by adjusting RGB values incrementally
    d.append(draw.Line(x1a + f * (x2a - x1a), y1a + f * (y2a - y1a), x1b + (1.0 - f) * (x2b - x1b), y1b + (1.0 - f) * (y2b - y1b), stroke=color, stroke_width=2, fill='none')) #string
    f = f + (1/(n-1)) #taking n-1 as the number of sections as the count of sections is one less than count of strings 

d #display drawing
```
