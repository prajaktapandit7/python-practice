```python
#Name: Prajakta Pandit
#Assignment 5: Author attribution that characterizes text documents based on the frequency of word use

def sort_dictionary( D ):                                         #function to sort dictionary based on frequency of words
    return sorted(D.items(), key=lambda x:x[1], reverse=True)               

def create_top_25 (DS):                                           #function to extract the top 25 most frequent words
    counter = 1
    top_25 = []     
    for i in DS:
        top_25.append(i)
        counter = counter + 1
        if counter > 25:                                          #condition to ensure only top 25 words are picked
            return top_25

def create_pct(word_ct_dict, N):
    for word, count in word_ct_dict.items():
        word_pct = count * (100/N)
        word_ct_dict[word] = word_pct
    return word_ct_dict

def print_dictionary( D, N ):                                        #function to print the dictionary
    D_pct = create_pct(D, N)
    DS = sort_dictionary( D_pct )                                     #the dictionary is first sorted
    top_25 = create_top_25(DS)                                    #the sorted dictionary is passed to the top_25 function
    print("\nPrinting Sorted Dictionary:\n")         
    for k in DS:
        print(k)
    return top_25

def strip_word( w ):                                              #function to strip word of punctuation and convert to all lower-case
    w = w.lower()
    w = ''.join([i for i in w if i.isalpha()])
    return( w )

def count_words( file_name, N = 0, word_ct_dict = {}):                     #function to count the words in dictionary
    myfile = open( file_name+'.txt', 'r', encoding="utf8" )       #opening file in read mode 
    for line in myfile:
        words = line.split(" ")                                   
        for w in words:
            w = strip_word( w )                                   #stripping punctuations
            if( len(w) > 0 ):
                if w in word_ct_dict: 
                    word_ct_dict[w] += 1                                   #incrementing frequency if word is present in dictionary
                else:
                    word_ct_dict[w] = 1                                    #adding new word if does not exist already
                N = N + 1                                     #incrementing counter
    myfile.close()
    return N, word_ct_dict

def build_dictionary( file_name ):                                #function to check parameter passed and determine how to process file
    word_ct_dict = {}
    N = 0
    if any(char.isdigit() for char in file_name):                 #condition to check if a particular filename has been passed eg. pk1, dr2 etc
        N, word_ct_dict = count_words( file_name, N, word_ct_dict)
    elif file_name == "pk" or file_name == "dr":                  #condition to handle case where 9 files for a particular author need to be processed
        for suffix in range(1, 10):
            N, word_ct_dict = count_words( file_name+str(suffix), N, word_ct_dict)
    elif file_name == "":                                         #condition to handle case where all 18 documents need to be processed
        for suffix in range(1, 10):
            N, word_ct_dict = count_words( "pk"+str(suffix), N, word_ct_dict)   
            N, word_ct_dict = count_words( "dr"+str(suffix), N, word_ct_dict)
    else:
        return print("\nPlease pass either pk, dr or no value")   #error handling condition
    
    return word_ct_dict, N                                                 #return top 25 words, dictionary, and number of words


            
print("\nBuilding a dictionary for all 18 documents:")
word_ct_dict, N = build_dictionary("")                                     #creating dictionary for all documents

print ("\nTotal number of words is:", N)                           
top_25 = print_dictionary( word_ct_dict, N )                                  #printing dictionary
```

```python
import numpy as np

def build_matrix(file_name, top_25):                          #function to create matrices
    
    F1 = np.zeros((9, 25), dtype=int)                         #initializing numpy matrices with 9 rows (for 9 docs) and 25 cols (for top 25 words)
    F2 = np.zeros((9, 25), dtype=int)                         #using data type integer

    for i in range(9):                                        #iterating through each row
        doc_name = file_name + str (i + 1)                    #generating file name
        doc_words, doc_N = build_dictionary(doc_name)         #building dictionary
        D_pct = create_pct(doc_words, doc_N)
        
        if file_name == "pk":                                 #case for author pk
            for j, (word, freq) in enumerate(top_25):         #looping through the top 25 words
                F1[i, j] = D_pct.get(word, 0)             #pulling frequency for the key 'word'; if does not exist use 0
        elif file_name == "dr":                               #case for author dr
            for j, (word, freq) in enumerate(top_25):
                F2[i, j] = D_pct.get(word, 0)
        else: return "Enter pk or dr!"                        #error handling condition

    if file_name == "pk":
        return F1
    else: return F2

F1 = build_matrix("pk", top_25)
print("\nMATRIX F1\n")
print(F1)
F2 = build_matrix("dr", top_25)
print("\nMATRIX F2\n")
print(F2)
```

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

F   = np.concatenate((F1,F2),axis=0)                         #use data from matrices F1 and F2 
pca = PCA(n_components=2)
Fp  = pca.fit(F).transform(F)
plt.scatter( Fp[0:9,0], Fp[0:9,1], color='b')                #generate plot
plt.scatter( Fp[9:18,0], Fp[9:18,1], color='r' )
```
