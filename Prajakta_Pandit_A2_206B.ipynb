```python
### Name: Prajakta Pandit
#Assignment 2: Implement a recursive binary search in which you search a sorted list of numbers for a specific number.
import random

def findBinary(L, key, left, right):
    
    if( len(L[left:right+1]) == 0 or left > right): #Condition to handle empty list or element not present
        return None

    else:
        midpoint  = int((left + right)/2 ) #calculating midpoint as average 
#        print("mid", midpoint)
#        print("key", key)
        if( L[midpoint] == key ):         #condition for locating element
            return midpoint
            
        elif( key  < L[midpoint] ):
#            print("less")
            return findBinary(L, key, left, midpoint-1)       #check left half of list
        else:
#            print("more")
            return findBinary(L, key, midpoint+1, right)      #check right half of list



user = "infinite"        #setting variable for infinite while loop

print("\n*************** Welcome to the Binary Search Tool ******************")

while (user == "infinite"):
    
    n = input("\nEnter the number of elements for your list: ")
    print("You entered: ", n)
    while n.isdigit() == 0:
        n = input("\nERROR! Please enter an integer value: ")      #error handling for number of elements 
    end = int(n)-1                                                 #defining right boundary
    random_list = []

    for i in range(int(n)):                                        #creating list of random integers
        x = random.randint(0, 1000)
        random_list.append(x)

    sorted_list = sorted(random_list)                              #sorting list
    print(sorted_list)

    print("\n*************** MENU *************** ")
    print("\nEnter an integer to search from the list")
    print("\nEnter ? to print the list")
    print("\nEnter q to quit")
    print("\n************************************")
    
    user = input("\nEnter your choice: ")                          #accepting user input
    print("\nYou entered: ", user)

    if user == "q":
        print("\nQuitting program.. Thanks for visiting!")         #exit condition
        break
    
    elif user == "?":                                              #print list
        print ("Your list is:", sorted_list)
        print("\n************************************")
        user = "infinite"
        
    else: 
        print("Location of element: ", findBinary( sorted_list, int(user), 0, end ))        #calling function to look for element
        print("\n************************************")
        user = "infinite"
    
```
