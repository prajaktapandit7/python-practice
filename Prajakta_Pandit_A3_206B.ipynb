```python
### Name: Prajakta Pandit
#Assignment 3: Write a Python function that performs bubble sort. Verify that your code works by 
#testing it on randomly generated lists and printing the lists before and after sorting to verify that the sorting is working properly.

import random

#function to swap element at index i and j in the list A
def swapElements(A,i):
    temp = A[i]
    A[i] = A[i+1]
    A[i+1] = temp
    
#function to sort by iteratively comparing each pair of adjacent items and swapping them
def BubbleSort(A):
    n = len(A)
        
    for j in range(n):                    #loop to check if more rounds are needed
        print ("\nPass ", j+1)
        status = "not swapped"
     
        for i in range(n-1-j):                #loop to move through the list (subtract j to ignore last elements which get sorted)
            
            if A[i] > A[i+1]:               #compare elements to check check if swap is needed
                swapElements(A,i)
                status = "swapped"
                print(A[i], " & ", A[i+1], "are swapped    ", ":", A)  #print current version of list
            else:    
                print(A[i], " & ", A[i+1], "are not swapped", ":", A)  #print current version of list
        
        if status == "not swapped":
            break;

def generate_list():
    random_list = []
    for i in range(random.randint(4, 8)):                                        #creating random length
        x = random.randint(0, 9)
        random_list.append(x)                                                    #creating list of random integers
    print("Original:    ", random_list)
    return random_list


for i in range(3):
    L1 = []
    L1 = generate_list()
    BubbleSort(L1)
    print("\nSorted list: ", L1)
    print("\n************************************************************************")

```

```python
# The run-time complexity of Bubble sort is O(n^2)
# The outer for loop is used to repeat the pass as long as the elements are not completely sorted .......... n
# The inner for loop is used to iterate through the list, compare the elements and perform swap if necessary ......... n-1-j
# The swap function has three lines ...... complexity 3, which we ignore as it is a constant
# We ignore the complexity of generating the list as the function is not a part of the algorithm
# In the best-case scenario, if the list is already sorted, the inner loop will run n-1 times and break out of outer loop due to no swaps .... n
# In the worst-case scenario, the inner loop will run comparisons n-1 times and if swaps occur every time, n-2, n-3 and so on times .... n(n-1)/2
# Ignoring constants, this comes up to n^2 - n, and we consider the term with largest power, which is O(n^2)
```
