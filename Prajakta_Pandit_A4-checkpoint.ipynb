```python
#Assignment 4: Python code that simulates bouncing and colliding balls
#Name: Prajakta Pandit
######################################################################

#importing relevant modules
import drawSvg as draw
import random
import math 

#defined global variables although not a recommended practice 
#as the assignment instructions asked us to only pass the variable i to the functions as an argument

#dimensions of rectangle
WINDOW_X = 400
WINDOW_Y = 300

#no. of balls
n = 8

#radius of ball
r = 10

#defining lists for x, y co-ordinate and velocity
x = []
y = []
vx = []
vy = []

########### function to draw ball
def draw_ball(i):

    d.append(draw.Circle(x[i], y[i], r, fill='yellow'))      

    return d

########### function that checks whether ball location is close to the sides of rectangle
#if true, reverses velocity to confine ball inside rectangle
def check_boundary(i):
    if x[i] >= 185 or x[i] <= -185:
        vx[i] = -1 * vx[i]
        
    if y[i] >= 135 or y[i] <= -135:
        vy[i] = -1 * vy[i]
        
    if (y[i] >= 135 or y[i] <= -135) and (x[i] >= 185 or x[i] <= -185):
        vx[i] = -1 * vx[i]
        vy[i] = -1 * vy[i]

    return vx[i], vy[i] 

########### function that changes location of the center co-ordinates of circle in increments
def update_position(i):

    vx[i], vy[i] = check_boundary(i)

#timestep default value is 1, value is updated in case of collision with other ball
    timestep = 1

#loop to check if balls are colliding by calculating distance between center coordinates
#and comparing with twice the radius    
    for j in range(i):
        if math.sqrt(((x[i]- x[j])*(x[i]- x[j])) + ((y[i]-y[j])*(y[i]- y[j]))) <= 2*r :
            vx[i] = -1 * vx[i] 
            vy[i] = -1 * vy[i]
            vx[j] = -1 * vx[j] 
            vy[j] = -1 * vy[j]
            timestep = 1.1
            vx[i], vy[i] = check_boundary(i)

#formula to increment x and y co-ordinates of the circle    
    x[i] = round(x[i] + timestep * vx[i])
    y[i] = round(y[i] + timestep * vy[i])
    
    draw_ball(i)

########### function to draw frames which updates the position for every ball
def draw_frame():
 
    for i in range(n):
        update_position(i)
        
    return d


#generating canvas for animation
with draw.animate_jupyter(draw_frame, delay=0.01) as anim:
    while( True ):
        d = draw.Drawing(WINDOW_X, WINDOW_Y, origin='center' ) # erase canvas

#assigning random values to x and y co-ordinates and velocities for n number of balls        
        for j in range(n):
        
            x.insert(j, round(random.uniform(-190, 190)))
            y.insert(j, round(random.uniform(-140, 140)))
    
            vx.insert(j, round(random.uniform(-5, 5)))
            vy.insert(j, round(random.uniform(-5, 5)))

#calling the frames to generate animation        
        while (True):
            d.append(draw.Rectangle(-WINDOW_X/2,-WINDOW_Y/2,WINDOW_X,WINDOW_Y, fill='black'))
            anim.draw_frame()

     
```
