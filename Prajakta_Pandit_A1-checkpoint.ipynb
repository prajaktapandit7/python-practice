```python
#NAME: PRAJAKTA PANDIT
#INFO 206A
#ASSIGNMENT 1.1: Write a Python program that prints out a logo containing your name

#define a function print_vert to print vertical patterns
#pass two variables: n defines number of characters, flag determines if vertical pattern should be repeated

def print_vert(n, flag):
    for x in range(n):
        if flag==1:
            print("P\tP")
        else:
            print("P")

#define a function print_horiz to print horizontal patterns
#pass one variable: n defines number of characters

def print_horiz(n):
    a = "" #define variable a so that the character is repeated n times
    for x in range(n):
        a = a + "P "
    print (a)    

#call functions to print horizontal and vertical characters to form the letter P for logo
print_horiz(4)
print_vert(3, 1)
print_horiz(4)
print_vert(2, 0)

#print name
print ("P rajakta Pandit")
```
